```python
# Each tuple defines a win if all cells in tuple contain players mark.
winning_combos = (
    (0, 1, 2), (3, 4, 5), (6, 7, 8),
    (0, 3, 6), (1, 4, 7), (2, 5, 8),
    (0, 4, 8), (2, 4, 6)
)

# Welcome and setup
print("Welcome to the Tic-Tac-Toe!\n")

# Display grid function
def print_grid():
    print("\n", " | ".join(grid[:3]))
    print("---+---+---")
    print("", " | ".join(grid[3:6]))
    print("---+---+---")
    print("", " | ".join(grid[6:]))

# Initial empty grid
grid = list("123456789")
print_grid()
print("")

# Player setup
player1 = input("Please enter name of player 1: ")
player1_symbol = input(f"Please enter the symbol of {player1}: ")

player2 = input("Please enter name of player 2: ")
while True:
    player2_symbol = input(f"Please enter the symbol of {player2} (not {player1_symbol}): ")
    if player2_symbol != player1_symbol:
        break
    print("Symbol must be different from player 1's symbol.")

# Set starting player
current_player = player1
current_symbol = player1_symbol

# Player move logic
def play(player, symbol):
    print_grid()
    while True:
        try:
            cell = int(input(f"{player}, enter your move (1-9): "))
            if str(cell) not in grid:
                raise ValueError
            grid[cell - 1] = symbol
            break
        except ValueError:
            print("Enter a valid number of an open cell.")

    # Check win condition
    for combo in winning_combos:
        if all(grid[cell] == symbol for cell in combo):
            return combo
    return None

# Game loop
for i in range(9):
    won = play(current_player, current_symbol)
    if won:
        print_grid()
        print(f"ðŸŽ‰ {current_player} wins!")
        break
    # Switch player
    if current_player == player1:
        current_player = player2
        current_symbol = player2_symbol
    else:
        current_player = player1
        current_symbol = player1_symbol
else:
    print_grid()
    print("Game ends in a draw.")
```
